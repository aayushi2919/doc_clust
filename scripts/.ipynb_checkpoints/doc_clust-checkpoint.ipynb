# Document Clustering 

A unique way of clustering docuemnts with the help of word vectors.

```python
from cluster import *
from gensim.models import Word2Vec
```

```python
# Supports Bengali, Hindi, Marathi, Tamil, Telugu
def get_example(lang):
    mdl = Word2Vec.load(r"../vectors/"+lang.lower()+".vec")
    d = get_doc(r"../data/"+lang.lower()+".csv.bz2")
    print("Document: ")
    print(d)
    
    # Get vectors 
    vecs = [mdl[x] for x in d.split() if x.strip() != ""]
    ci = crp(vecs)
    clusters = []
    for clst in ci:
        clusters.append([d.split(" ")[widx] for widx in clst])
    print("Clusters: "+str(len(clusters)))
    for c in clusters:
        print(list(set(c)))
    return clusters

def cluster_text(lang, text):
    mdl = Word2Vec.load(r"../vectors/"+lang.lower()+".vec")
    print("Documnt: ")
    print(text)
    # Get vectors 
    vecs = [mdl[x] for x in text.split() if x.strip() != ""]
    ci = crp(vecs)
    clusters = []
    for clst in ci:
        clusters.append([text.split(" ")[widx] for widx in clst])
    print("Clusters: "+str(len(clusters)))
    for c in clusters:
        print(list(set(c)))
    return clusters
```
